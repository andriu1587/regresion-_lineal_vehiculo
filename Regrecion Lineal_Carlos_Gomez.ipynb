```python
# pUNTO 3 SELECCIONAR LAS VARIABLES 
# convertir los datos en un dataframe
df = pd.DataFrame(datos)
```

```python
# MODELO DE REGRESION SIMPLE
# VARIABLE INDEPENDIENTE(MARCA) VARIABLE DEPENDIENTE (PRECIO)
x = df[['Year']]
y = df [['Price']]

# creacion del modelo de regresion lineal simple
modelo = LinearRegression()

# ajustar el modelo a los datos
modelo.fit(x,y)

# coeficientes de la regresion
beta_0 = modelo.intercept_
beta_1 = modelo.coef_

# mostrar los coeficientes
print(f"intercepto (Price,β0): {beta_0}")
print(f"coeficiente (Model,β1): {beta_1}")

# VISIALIZAR LA REGRESION LINEAL
plt.scatter(df['Year'], y, color='green' , label= 'datos reales')
plt.plot(df['Year'], modelo.predict(x), color='blue' ,label= 'linea de regresion')
plt.title('regresion lineal: precio VS año')
plt.xlabel('año de fabricacion vehiculo')
plt.ylabel('precio')
plt.legend()
plt.show()
```
